```python
# basic lenear regression   
# Path: 0x040-AI/064-sciki-learn/020-basic-linear-regression.ipynb
# lenear regresssion means that the output is a linear function of the input i.e. y = mx + c 
# where m is the slope and c is the intercept
# y = 2x + noise

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# create random data
rng = np.random.RandomState(1) # seed the random number generator to get the same result
x = 10 * rng.rand(50) # 50 random numbers between 0 and 10
y = 2 * x - 1 + rng.randn(50)
#y = 2 * x*x - 1 + rng.randn(50) 
#y = 2 * x*x*x - 1 + rng.randn(50)
plt.scatter(x, y)

# fit the model
model = LinearRegression(fit_intercept=True)
model.fit(x[:, np.newaxis], y)

# predict
xfit = np.linspace(0, 10, 1000) # 1000 points between 0 and 10
yfit = model.predict(xfit[:, np.newaxis]) # np.newaxis is used to add a new axis to the array xfit
# axis in array represent the dimension of the array exapmle: 1D array has 1 axis, 2D array has 2 axis, 3D array has 3 axis
#

# plot the result
plt.scatter(x, y)
plt.plot(xfit, yfit)
plt.show()



```

```python
# basic polynomial regression
# Path: 0x040-AI/064-sciki-learn/030-basic-polynomial-regression.ipynb
# polynomial regression means that the output is a polynomial function of the input i.e. y = ax^2 + bx + c
# where a, b and c are the coefficients
# y = 2x^2 + noise

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# create random data
rng = np.random.RandomState(1) # seed the random number generator to get the same result
x = 10 * rng.rand(50) # 50 random numbers between 0 and 10
y = x+2 * x*x - 1 + rng.randn(50)
#y = x+2 * x*x*x*x - 1 + rng.randn(50)
#y = x+2 *x - 1 + rng.randn(50)
plt.scatter(x, y)

# fit the model
poly = PolynomialFeatures(degree=2, include_bias=False) # PolynomialFeatures is used to create polynomial features i.e. x^2, x^3, x^4, etc
x2 = poly.fit_transform(x[:, np.newaxis]) # fit_transform is used to fit the model and transform the data at the same time
# x2 is the transformed data with polynomial features i.e. x^2, x^3, x^4, etc
# x2 is a 2D array with 2 columns and 50 rows (50 data points) which represent the x and x^2 values of the data points
#print (x2.shape)
#print (x)
#print (x2)

model = LinearRegression(fit_intercept=True)
model.fit(x2, y)

# predict
xfit = np.linspace(0, 10, 1000) # 1000 points between 0 and 10
yfit = model.predict(poly.fit_transform(xfit[:, np.newaxis])) # np.newaxis is used to add a new axis to the array xfit
# axis in array represent the dimension of the array exapmle: 1D array has 1 axis, 2D array has 2 axis, 3D array has 3 axis

# plot the result
plt.scatter(x, y)
plt.plot(xfit, yfit)
plt.show()

```
