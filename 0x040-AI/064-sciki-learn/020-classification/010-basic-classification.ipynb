```python
# in case of classification, the target is a class label
# in case of regression, the target is a real value
# in case of clustering, the target is not defined

# example digit classification problem with the digits dataset from scikit-learn 
# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_digits.html
# data can be any digit from 0 to 9
# target is the digit itself
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# load the digits dataset
digits = load_digits()
#print (digits)

# print the shape of the data
print(digits.data.shape)

# print the shape of the target
print(digits.target.shape)

# print the first 10 images
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(digits.images[i], cmap=plt.cm.gray_r, interpolation='nearest')
    plt.axis('off')
plt.show()
```

```python
# train the model 
# split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.4, random_state=0)

# create the model
model = LogisticRegression( solver='lbfgs', multi_class='auto', max_iter=10000)

# fit the model
model.fit(X_train, y_train)

# predict the target
y_pred = model.predict(X_test)

# print the accuracy
print(accuracy_score(y_test, y_pred))

# print the first 10 predictions
print(y_pred[:10])


# function to predict 
def predict_digit(i):

    # print the first 10 targets
    print(y_test[i])

    
    # reshape the first image
    image = X_test[i].reshape(8, 8)
    # display the image
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    plt.axis('off')
    plt.show()

# call the predict function for  0 to 1 in loop 
for i in range(3):
    predict_digit(i)   










```
