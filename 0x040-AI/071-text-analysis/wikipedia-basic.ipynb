```python
# import wikipedia liberary 
import wikipedia
```

```python
# get the topic of interest and print conmtent 
topic = "canadian independence "
#print(wikipedia.summary(topic))

# print the title of the page
print(wikipedia.page(topic).title)

# print the url of the page
print(wikipedia.page(topic).url)

# print the content of the page by paragraph
for paragraph in wikipedia.page(topic).content.split("\n"):
    # trim the paragraph
    paragraph = paragraph.strip()
    # get the length of the paragraph
    length = len(paragraph) 
    # print the paragraph if it is not empty
    if length > 0:
        print(length)
        print(paragraph)    
print()
#print(wikipedia.page(topic).content)

```
